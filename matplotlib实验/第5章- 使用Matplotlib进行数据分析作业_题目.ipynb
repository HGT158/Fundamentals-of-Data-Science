### Step3. 绘制条形图

### Step4. 绘制饼图

```python
# your code
survived_matrix = pd.crosstab(data1['Survived'], data1['Age'])

survived_matrix
```

```python
survived_counts = data1.groupby('Age')['Survived'].value_counts().unstack()

adult_survived_ratio = survived_counts.loc['Adult']['Yes'] / (survived_counts.loc['Adult']['Yes'] + survived_counts.loc['Adult']['No'])
child_survived_ratio = survived_counts.loc['Child']['Yes'] / (survived_counts.loc['Child']['Yes'] + survived_counts.loc['Child']['No'])
fig, axes = plt.subplots(1, 2, figsize=(10, 5))


axes[0].pie([adult_survived_ratio, 1 - adult_survived_ratio], labels=['Yes', 'No'], autopct='%1.2f%%')
axes[0].set_title('Adult Survival Rate')
axes[0].axis('equal')


axes[1].pie([child_survived_ratio, 1 - child_survived_ratio], labels=['Yes', 'No'], autopct='%1.2f%%')
axes[1].set_title('Children Survival Rate')
axes[1].axis('equal')

plt.tight_layout()
plt.show()
```

### Step9. 绘制每加仑油行驶的英里数（mpg）、总马力（hp）和汽车自重（wt）3个变量的气泡图

```python
x = data3['mpg']
y = data3['hp']
size_var = data3['wt']

wt_min, wt_max = size_var.min(), size_var.max()
size_pt2 = 80 + (size_var - wt_min) / (wt_max - wt_min) * (1200 - 80)

cmap = plt.cm.cividis

fig, ax = plt.subplots(figsize=(7, 5.5))
sc = ax.scatter(
    x, y,
    s=size_pt2,
    c=size_var,
    cmap=cmap,
    alpha=0.85,
    edgecolors='white',
    linewidths=0.7
)

ax.set_xlabel('每加仑油行驶里程 (mpg)')
ax.set_ylabel('总马力 (hp)')
ax.set_title('MPG vs HP 气泡图（气泡大小与颜色表示自重 wt）')

cbar = plt.colorbar(sc, ax=ax, pad=0.02)
cbar.set_label('自重 wt（千磅）')

ax.grid(alpha=0.25, linestyle='--')

plt.tight_layout()
plt.show()
```

```python
GDP = pd.read_excel('./世界各国GDP（1960-2019）.xls')
GDP.drop(columns='2020', axis=1, inplace=True)
GDP.head()
```

### Step12 根据国家真实GDP数据画图：画出中国1960年至2015年的GDP曲线图，取值可以用切片(iloc[0,4:60])

```python
# your code
name_cols = {'country name','国家名称','国家','country'}
code_cols = {'country code','国家代码','代码','code'}
cn = None

code_col = next((c for c in GDP.columns if str(c).strip().lower() in code_cols), None)
if code_col is not None:
    cn = GDP[GDP[code_col].astype(str).str.upper().eq('CHN')]
if (cn is None or cn.empty):
    name_col = next((c for c in GDP.columns if str(c).strip().lower() in name_cols), None)
    if name_col is not None:
        cn = GDP[GDP[name_col].astype(str).isin(['China','中国','中华人民共和国'])]

row = (cn.iloc[0] if cn is not None and not cn.empty else GDP.iloc[0])

years = range(1960, 2016)
x, y = [], []
for yr in years:
    col = yr if yr in GDP.columns else (str(yr) if str(yr) in GDP.columns else next((c for c in GDP.columns if str(c).startswith(str(yr))), None))
    if col is not None:
        val = pd.to_numeric(pd.Series([row[col]]), errors='coerce').iloc[0]
        if pd.notna(val):
            x.append(yr); y.append(val)

plt.figure(figsize=(6,4.5))
plt.plot(x, y, lw=2)
plt.title('中国GDP变化趋势')
plt.xlabel('年份')
plt.ylabel('GDP')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

### Step13 根据国家真实GDP数据画图：画出以所有国家2015年GDP为小于百亿，小于千亿，小于万亿和大于万亿的国家占比的饼图，可以将NaN填为0

```python
# 定位2015列
if 2015 in GDP.columns:
    col2015 = 2015
elif '2015' in GDP.columns:
    col2015 = '2015'
else:
    col2015 = next((c for c in GDP.columns if str(c).startswith('2015')), None)
if col2015 is None:
    raise ValueError('未找到2015列，请检查Excel列名')

vals = pd.to_numeric(GDP[col2015], errors='coerce').fillna(0)

B, T = 1e9, 1e12
base_labels = ['<100亿', '<1000亿', '<1万亿', '≥1万亿']

# 原始边界
bins = np.array([-np.inf, 100*B, 1000*B, 1*T, np.inf], dtype=float)

# 去重（保持顺序）
uniq_bins = np.unique(bins)
if len(uniq_bins) != len(bins):
    # 根据实际段数裁剪/重建标签，避免数量不匹配
    k = len(uniq_bins) - 1
    labels = base_labels[:k]
else:
    labels = base_labels

cats = pd.cut(vals, bins=uniq_bins, labels=labels, right=True)
counts = pd.Series(cats).value_counts().reindex(labels).fillna(0).astype(int)

plt.figure(figsize=(6,6))
plt.pie(counts, labels=[f'{k} ({v})' for k,v in zip(labels, counts)], autopct='%.1f%%', startangle=90)
plt.title('各国2015年GDP规模占比')
plt.tight_layout(); plt.show()
```

### Step14 根据国家真实GDP数据画图：画出日本与韩国的折线图，两条折线画在一张图中

```python
name_col = next((c for c in GDP.columns if str(c).strip().lower() in {'country name','国家名称','国家','country'}), None)
if name_col is None:
    raise ValueError('未找到国家名称列（如 国家/国家名称/Country Name）。')
jp_row = GDP[GDP[name_col].astype(str).eq('日本')].iloc[0]
kr_row = GDP[GDP[name_col].astype(str).eq('大韩民国')].iloc[0]

# 收集年份与数值
years, jp_vals, kr_vals = [], [], []
for y in range(1960, 2020):
    col = y if y in GDP.columns else (str(y) if str(y) in GDP.columns else next((c for c in GDP.columns if str(c).startswith(str(y))), None))
    if col is None: 
        continue
    jv = pd.to_numeric(pd.Series([jp_row[col]]), errors='coerce').iloc[0]
    kv = pd.to_numeric(pd.Series([kr_row[col]]), errors='coerce').iloc[0]
    if pd.notna(jv) or pd.notna(kv):
        years.append(y)
        jp_vals.append(jv if pd.notna(jv) else None)
        kr_vals.append(kv if pd.notna(kv) else None)

# 绘图
step = 1 # 标记点间隔
plt.figure(figsize=(7,4.8))
plt.plot(years, jp_vals, label='日本', lw=1.8, marker='o', ms=5, markevery=step)
plt.plot(years, kr_vals, label='大韩民国', lw=1.8, marker='o', ms=5,  markevery=step)
plt.title('日本与韩国GDP')
plt.xlabel('年份'); plt.ylabel('GDP')
plt.grid(True, alpha=0.3)
plt.legend(loc="upper left")
plt.tight_layout()
plt.show()
```

### Step15 根据国家真实GDP数据画图：画出日本与韩国的面积图，画在一张图中

```python
#your code
plt.figure(figsize=(8,5))
plt.stackplot(years, jp_vals, kr_vals, labels=['日本', '韩国'], alpha=0.7)
plt.title('日本与韩国GDP面积图')
plt.xlabel('年份'); plt.ylabel('GDP')
plt.legend(loc='upper left')
plt.grid(True, alpha=0.25)
plt.tight_layout()
plt.show()
```
